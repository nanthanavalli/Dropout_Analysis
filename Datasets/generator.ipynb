```python
import pandas as pd

# Load the sample dataset into a DataFrame
data = pd.read_csv('C:\\Users\\NaveenAkash\\Documents\\Code\\MachineLearning\\fake_dataset\\Untitled Folder 1\\ALFRED HIGH SCHOOL.csv')

# Calculate the target number of students per section (60 students)
target_students_per_section = 60

# Initialize a new DataFrame to store the updated data
updated_data = pd.DataFrame(columns=data.columns)

# Iterate through each class and section
for grade in range(1, 13):
    for section in ['A', 'B', 'C', 'D']:
        # Filter students from the current class and section
        filtered_data = data[(data['class'] == grade) & (data['section'] == section)]

        # Shuffle the filtered data randomly
        filtered_data = filtered_data.sample(frac=1, random_state=42)

        # Calculate the number of students in the current section
        num_students_in_section = len(filtered_data)

        # Distribute students to sections ensuring no more than 60 in each
        while num_students_in_section > 0:
            students_to_move = min(target_students_per_section - num_students_in_section, num_students_in_section)
            students_to_move = max(students_to_move, 0)  # Ensure no negative values

            # Add the selected students to the updated_data DataFrame
            updated_data = pd.concat([updated_data, filtered_data.iloc[:students_to_move]])

            # Remove the selected students from the filtered_data
            filtered_data = filtered_data.iloc[students_to_move:]

            num_students_in_section -= students_to_move

# Save the updated data to a new CSV file
updated_data.to_csv('C:\\Users\\NaveenAkash\\Documents\\Code\\MachineLearning\\fake_dataset\\Untitled Folder 1\\ALFRED HIGH SCHOOL updated_dataset.csv', index=False)
```
