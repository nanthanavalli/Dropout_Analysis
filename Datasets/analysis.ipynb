```python
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv("sample_dataset.csv", encoding='ISO-8859-1')

# Calculate the total number of students
total_students = len(df)

# Calculate the number of dropped out students
dropped_out_students = len(df[df["discontinue"] == "Yes"])

# Calculate the percentage of dropped out students
dropout_percentage = (dropped_out_students / total_students) * 100

# Define a function to calculate the percentage of field values corresponding to dropped out students
def calculate_field_percentage(field_name):
    # Filter the DataFrame for dropped out students
    dropout_students_df = df[df["discontinue"] == "Yes"]
    # Calculate the percentage of field values for dropped out students
    field_percentages = dropout_students_df[field_name].value_counts(normalize=True) * 100
    return field_percentages

# List of fields to analyze
fields_to_analyze = ["gender", "community", "class", "city", "district", "religion", "discontinueReason","schoolType", "school","mediumOfInstruction","motherTongue","fathersOccupation","mothersOccupation","fathersEducation","mothersEducation"]

# Display total number of students, dropped out students, and dropout percentage
print(f"Total number of students: {total_students}")
print(f"Number of dropped out students: {dropped_out_students}")
print(f"Percentage of dropped out students: {dropout_percentage:.2f}%\n")

# Calculate and display field percentages for dropped out students
for field in fields_to_analyze:
    print(f"Field percentages for {field} (corresponding to dropped out students):")
    field_percentages = calculate_field_percentage(field)
    print(field_percentages)
    print("\n")
```
